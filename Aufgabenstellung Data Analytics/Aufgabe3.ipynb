```python
import pandas as pd
import seaborn as sns
```

```python
package_data['timestamp'] =  pd.to_datetime(package_data['timestamp'])
```

```python
package_data['date_column'] = pd.to_datetime(package_data['timestamp']).dt.date
```

```python
g = sns.lineplot(x="date_column", y="good_packs", data=package_data)
g.set_xticklabels(g.get_xticklabels(), rotation=45)
```
